```python
# import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/datasets/API_SP.POP.TOTL_DS2_en_csv_v2_606680.csv"
# on_bad_lines='skip'
df=pd.read_csv(path,skiprows=4)
df.head(5)
population=df['2023'].head(10)
countries=df['Country Name'].head(10)
print(population)
print(countries)

positions=range(len(countries))
plt.figure(figsize=(20, 10))
plt.bar(positions,population)
plt.xticks(positions,countries)
plt.xlabel('Countries')
plt.ylabel('Population in 2023')
plt.show()
```
